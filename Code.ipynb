```python
def yolo_filter_boxes(boxes, box_confidence, box_class_probs, threshold=0.6):
    box_scores = box_confidence * box_class_probs
    box_classes = tf.math.argmax(box_scores, axis=-1)
    box_class_scores = tf.math.reduce_max(box_scores, axis=-1)
    filtering_mask = box_class_scores >= threshold
    # Check if filtering_mask is empty and handle the case
    if not tf.math.reduce_any(filtering_mask):  # Check if any element is True
        return tf.constant([]), tf.constant([]), tf.constant([])  # Return empty tensors

    scores = tf.boolean_mask(box_class_scores, filtering_mask)
    boxes = tf.boolean_mask(boxes, filtering_mask)
    classes = tf.boolean_mask(box_classes, filtering_mask)
    return scores, boxes, classes

def iou(box1, box2):
    (box1_x1, box1_y1, box1_x2, box1_y2) = box1
    (box2_x1, box2_y1, box2_x2, box2_y2) = box2

    xi1 = max(box1_x1, box2_x1)
    yi1 = max(box1_y1, box2_y1)
    xi2 = min(box1_x2, box2_x2)
    yi2 = min(box1_y2, box2_y2)
    inter_width = max(0, xi2 - xi1)
    inter_height = max(0, yi2 - yi1)
    inter_area = inter_width * inter_height

    box1_area = (box1_y2 - box1_y1) * (box1_x2 - box1_x1)
    box2_area = (box2_y2 - box2_y1) * (box2_x2 - box2_x1)
    union_area = box1_area + box2_area - inter_area

    return inter_area / union_area if union_area > 0 else 0

def yolo_non_max_suppression(scores, boxes, classes, max_boxes=10, iou_threshold=0.5):
    max_boxes_tensor = tf.Variable(max_boxes, dtype='int32')
    nms_indices = tf.image.non_max_suppression(boxes, scores, max_boxes_tensor, iou_threshold)
    scores = tf.gather(scores, nms_indices)
    boxes = tf.gather(boxes, nms_indices)
    classes = tf.gather(classes, nms_indices)
    return scores, boxes, classes
```

```python
import os
import cv2
import json  # or other libraries depending on your annotation format
import numpy as np

def load_validation_data(validation_folder):
    """Load images and their corresponding annotations from a validation folder.

    Arguments:
    validation_folder -- path to the validation folder containing 'images' and 'annotations' directories.

    Returns:
    validation_data -- list of tuples containing (image, ground_truths).
    """

    images_folder = os.path.join(validation_folder, "images")
    annotations_folder = os.path.join(validation_folder, "labels")

    validation_data = []

    # Iterate over all images in the images folder
    for image_filename in os.listdir(images_folder):
        if image_filename.endswith(('.jpg', '.jpeg', '.png', '.webp')):  # Add other image formats as needed
            # Load the image
            image_path = os.path.join(images_folder, image_filename)
            image = cv2.imread(image_path)

            # Load the corresponding annotation file
            annotation_filename = image_filename.replace('.jpg', '.txt').replace('.jpeg', '.txt').replace('.png', '.txt').replace('.webp', '.txt')
            annotation_path = os.path.join(annotations_folder, annotation_filename)

            if os.path.exists(annotation_path):
                # Load the annotations from the text file
                ground_truths = []
                with open(annotation_path, 'r') as f:
                    for line in f:
                        # Split the line into components
                        parts = line.strip().split()
                        class_id = int(parts[0])  # First value is the class label
                        x_center = float(parts[1])  # Second value is x center
                        y_center = float(parts[2])  # Third value is y center
                        width = float(parts[3])  # Fourth value is width
                        height = float(parts[4])  # Fifth value is height

                        # Convert (x_center, y_center, width, height) to (x1, y1, x2, y2)
                        x1 = x_center - width / 2
                        y1 = y_center - height / 2
                        x2 = x_center + width / 2
                        y2 = y_center + height / 2

                        # Append the bounding box and class ID as a tuple
                        ground_truths.append(((x1, y1, x2, y2), class_id))

                # Append the image and ground truths to validation_data
                validation_data.append((image, ground_truths))

    return validation_data

# Example usage
validation_folder = "/content/gdrive/MyDrive/Robot_parts_question/Dataset/validation"  # Change this to your actual validation folder path
validation_data = load_validation_data(validation_folder)
```

```python
def evaluate_model(model, validation_data):
    """Evaluate the model on the validation data.

    Arguments:
    model -- the YOLO model to evaluate.
    validation_data -- a list of tuples containing (image, ground_truths).

    Returns:
    predictions -- list of predictions from the model.
    ground_truths -- list of ground truths from the validation data.
    """
    predictions = []
    ground_truths = []

    for image, gt_boxes in validation_data:
        # Preprocess the image for the model
        # Ensure image is a tensor and has the right shape (1, C, H, W)
        # Resize image to be divisible by stride 32
        height, width, _ = image.shape
        new_height = (height // 32) * 32
        new_width = (width // 32) * 32
        resized_image = cv2.resize(image, (new_width, new_height))  # Resize using OpenCV

        image_tensor = torch.tensor(resized_image).permute(2, 0, 1).unsqueeze(0).float()/255  # Change to (1, 3, H, W)

        # Move tensor to GPU if available
        if torch.cuda.is_available():
            image_tensor = image_tensor.to('cuda')

        with torch.no_grad():  # Disable gradient calculation for evaluation
            results = model(image_tensor)  # Get results

        # Access boxes, scores, and classes using Ultralytics API
        for result in results:
            if result is not None and len(result):
                boxes = result.boxes.xyxy.cpu().numpy()  # Get bounding boxes
                scores = result.boxes.conf.cpu().numpy()  # Get confidence scores
                classes = result.boxes.cls.cpu().numpy()  # Get class IDs

                # Store predictions
                predictions.append((boxes, scores, classes))
            else:
                predictions.append(([], [], []))  # No detections

        # Store ground truths
        ground_truths.append(gt_boxes)

    return predictions, ground_truths

# Example usage
predictions, ground_truths = evaluate_model(model1, validation_data)

print(predictions)
print(ground_truths)
```
